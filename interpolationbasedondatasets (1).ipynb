```python
import random
import time

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            return -1, probes

        pos = low + int(
            ((key - arr[low]) * (high - low)) /
            (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        if arr[pos] == key:
            return pos, probes

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


def get_dataset(choice):
    if choice == 1:  # Uniform Dataset
        method = int(input(
            "\n1. Enter elements manually\n2. Generate randomly\nChoose: "
        ))

        if method == 1:
            arr = list(map(int, input("Enter sorted elements: ").split()))
            arr.sort()
        else:
            n = int(input("Enter number of elements: "))
            start = int(input("Enter starting value: "))
            gap = int(input("Enter uniform gap: "))
            arr = [start + i * gap for i in range(n)]

    else:  # Non-Uniform Dataset
        method = int(input(
            "\n1. Enter elements manually\n2. Generate randomly\nChoose: "
        ))

        if method == 1:
            arr = list(map(int, input("Enter sorted elements: ").split()))
            arr.sort()
        else:
            n = int(input("Enter number of elements: "))
            arr = sorted(random.randint(1, n * n) for _ in range(n))

    return arr


print("===== INTERPOLATION SEARCH =====")

print("\nChoose Dataset Type")
print("1. Uniformly Distributed Dataset")
print("2. Non-Uniformly Distributed Dataset")

choice = int(input("Enter choice: "))

dataset = get_dataset(choice)

print("\nDataset:")
print(dataset)

key = int(input("\nEnter search key: "))

start_time = time.perf_counter()

index, probes = interpolation_search(dataset, key)

end_time = time.perf_counter()

execution_time = (end_time - start_time) * 1000

print("\n===== RESULT =====")

if index != -1:
    print(f"Key {key} found at index {index}")
else:
    print(f"Key {key} not found")

print(f"Number of Probes/Comparisons : {probes}")
print(f"Execution Time : {execution_time:.6f} ms")

print("\n===== PERFORMANCE NOTE =====")
if choice == 1:
    print("Uniform dataset selected.")
    print("Interpolation Search is generally very efficient on uniformly distributed data.")
else:
    print("Non-uniform dataset selected.")
    print("Interpolation Search may require more probes and perform less efficiently.")
```
